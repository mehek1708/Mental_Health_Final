### **3. Theoritical Background**

This project uses a mix of unsupervised and supervised machine learning techniques to explore and model mental health inequality at a global scale. Below is a brief explanation of each technique used, along with its purpose in this analysis.

- ***Principal Component Analysis (PCA):***  
PCA is a dimensionality reduction technique that transforms a dataset with many features into a smaller set of uncorrelated variables called *principal components*. These components capture the maximum variance in the data, helping us simplify complex datasets without losing important information.  
In this project, PCA is used to reduce the number of features (like access indicators and professional availability) and visualize which factors explain the most variability between countries. It also helps with visualizing inherent structure in the dataset by projecting it into 2D space.

- ***Singular Value Decomposition (SVD):***  
SVD is another technique for matrix factorization that breaks down a matrix into three parts—U, Σ, and Vᵗ. It's mathematically similar to PCA and is used to understand underlying data structure, especially when dealing with missing values.  
In our case, SVD complements PCA and helps confirm how much structure (and noise) exists in the dataset, giving us insight into the data’s complexity.

- ***Matrix Completion:***  
Real-world data often comes with missing values. Matrix completion helps estimate those missing entries by assuming the data lies on a low-rank structure (i.e., the full dataset can be explained by a smaller set of patterns).  
Here, it is used to fill in incomplete entries in access-to-care indicators, workforce availability, and GDP values without biasing the results by dropping rows or columns.

- ***K-Means Clustering:***  
K-means is an unsupervised learning algorithm that groups data points into “k” clusters based on similarity. It tries to minimize the distance between points within a cluster and the cluster's centroid.  
In this project, k-means helps identify groups of countries with similar mental health infrastructure. This clustering gives us insight into which countries are performing similarly (either well or poorly) in terms of access and policy.

- ***Hierarchical Clustering:***  
Hierarchical clustering builds a tree-like structure (dendrogram) to show how countries group together based on their features. Unlike k-means, it doesn’t require pre-selecting the number of clusters and lets us explore structure at different granularity levels.  
This is useful for identifying not just flat clusters, but nested relationships between countries, such as whether high-income countries cluster separately from middle-income ones.

- ***Supervised Learning (Logistic Regression, Random Forest, Decision Trees):***  
Supervised models were used to predict the likelihood that a country has key mental health infrastructure based on other features (e.g., workforce size, access indicators). Logistic Regression provided a simple baseline, while Random Forests and Decision Trees helped identify which features were most predictive and gave more flexible model structures.

### **4. Methodology**

This project involved bringing together multiple datasets from different sources to explore patterns in global mental health infrastructure using machine learning. Because no single dataset contained everything we needed, one of the most important parts of this project was data cleaning, merging, and handling time inconsistencies.

##### **41. Data Sources and Structure**  
- ***Mental Health Disorder DALYs (Our World in Data):***  
This dataset provides the burden of disease (in terms of disability-adjusted life years) due to mental health and substance use disorders. The available data mostly spans 2019–2021.

- ***Access to Care and Infrastructure (WHO Mental Health Atlas):***  
This includes policy presence, number of mental health professionals, and access-related indicators. The most complete version of this dataset covers the years 2011–2014.

- ***GDP per Country (World Bank):***  
Included GDP to represent a country’s economic capacity, with data ranging from 2011 to 2021.

- ***Psychiatrists per Capita (WHO Atlas):***  
This variable came from similar infrastructure reports but had some variation in its available years.

##### **4.2 Time Range Differences & Approach**  
One of the biggest challenges was the inconsistent time coverage across datasets. While infrastructure data like policy and workforce availability was mostly from 2011–2014, the disorder burden (DALYs) was from 2019–2021. Similarly, GDP data spanned a wider range (2011–2021).

After careful consideration, I still decided to move forward with the combined dataset for the following reasons:  
1. ***Limited data availability:***   
   The only complete global infrastructure data was from 2011–14. No newer dataset with equal breadth was available.

2. ***Stability assumption:***   
   Mental health infrastructure (like national policies and workforce) doesn’t usually change rapidly. So, using disorder burden from later years still allows for meaningful interpretation, assuming general stability in mental health systems over time.

3. ***Analysis goals:***   
   Since the objective was not to forecast or establish time-dependent causality, but to explore structural relationships between countries, the small time gap was considered acceptable for an unsupervised analysis.

This mismatch throughout the analysis was acknowledged and hence, making any time-sensitive conclusions was refrained.

##### **4.3 Data Cleaning and Preprocessing**
- The basic cleaning of the raw datafiles was done in ***'basic_cleaning.ipynb'***.
  
- Missing values were handled using SimpleImputer with mean or most-frequent strategies depending on the variable type.

- All variables were normalized or scaled before applying PCA and clustering algorithms to ensure that larger-value features (like GDP) didn’t dominate the results.

- Categorical values (e.g., Yes/No for policy presence) were encoded numerically for compatibility with machine learning models.

##### **4.4 Analysis Workflow**
- PCA and SVD were used for dimensionality reduction and to visualize variance explained by each component.

- Matrix completion helped fill in missing entries while preserving overall structure.

- K-means and hierarchical clustering were used to group countries by similarity in access and infrastructure.

- Supervised models (Logistic Regression, Random Forest, and Decision Trees) were used to predict key features such as policy presence based on other available infrastructure data.

```python
# Extract latest available year for each access indicator

# Policies
latest_policies = policies_clean.sort_values('Year').drop_duplicates('Country_Code', keep='last')[['Country_Code', 'Has_MH_Policy']]

# Services
latest_services = services_clean.sort_values('Year').drop_duplicates('Country_Code', keep='last')[['Country_Code', 'Has_MH_Services']]

# Legislation
latest_legislation = legislation_clean.sort_values('Year').drop_duplicates('Country_Code', keep='last')[['Country_Code', 'Has_MH_Legislation']]

# Merge them all
latest_access_indicators = latest_policies.merge(latest_services, on='Country_Code', how='outer')
latest_access_indicators = latest_access_indicators.merge(latest_legislation, on='Country_Code', how='outer')

# Add country name by merging with country code reference
latest_access_indicators = latest_access_indicators.merge(
    country_code_clean[['Code', 'Title']],
    left_on='Country_Code',
    right_on='Code',
    how='left'
)

# Rename for clarity
latest_access_indicators.rename(columns={'Title': 'Country_Name'}, inplace=True)

# Reorder columns
latest_access_indicators = latest_access_indicators[['Country_Name', 'Country_Code', 'Has_MH_Policy', 'Has_MH_Services', 'Has_MH_Legislation']]

# Preview
display(latest_access_indicators.head())
```

```python
# Convert Yes/No to binary values (1 for Yes, 0 for No)
latest_access_indicators.replace({
    'Has_MH_Policy': {'Yes': 1, 'No': 0},
    'Has_MH_Services': {'Yes': 1, 'No': 0},
    'Has_MH_Legislation': {'Yes': 1, 'No': 0}
}, inplace=True)

# Preview
print(" Access Indicators (Numeric Format):")
display(latest_access_indicators.head())
```

```python
# Grouping and pivoting to get average DALY per disorder per country
all_dalys = pd.concat([disorders_daly, substance_daly], ignore_index=True)

daly_summary = all_dalys.groupby(['Country_Name', 'Country_Code', 'Disorder_Name'])['Disorder_Rate'].mean().reset_index()

daly_pivot = daly_summary.pivot(index=['Country_Name', 'Country_Code'], columns='Disorder_Name', values='Disorder_Rate').reset_index()

# Merging with access indicators 
model_data = daly_pivot.merge(latest_access_indicators, on='Country_Code', how='inner')

# Dropping duplicate country name columns if they exist
if 'Country_Name_y' in model_data.columns:
    model_data.drop(columns='Country_Name_y', inplace=True)
    model_data.rename(columns={'Country_Name_x': 'Country_Name'}, inplace=True)

# Dropping rows with missing values
model_data.dropna(inplace=True)

# Reordering columns for clarity
cols = ['Country_Name', 'Country_Code'] + [col for col in model_data.columns if col not in ['Country_Name', 'Country_Code']]
model_data = model_data[cols]

# Preview
print("Final dataset for modeling:")
display(model_data.head())
```

```python
# Matrix Completion

# Purpose: Fill missing data points while also preserving the structure.

# Remove identifiers and impute missing values
features = model_data.drop(columns=['Country_Name', 'Country_Code'])
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(features)
```

```python
# Feature Scaling

# Purpose: Standardize features for PCA and clustering.

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)
```

```python
# Dimensionality reduction for PCA

pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
# PCA Scree Plot

# Purpose: Explained Variance by Principal Components

explained_var = pca.explained_variance_ratio_
components = np.arange(1, len(explained_var) + 1)
cumulative_var = np.cumsum(explained_var)

plt.figure(figsize=(10, 6))
plt.bar(components, explained_var, alpha=0.7, label='Individual Explained Variance', color='skyblue')
plt.plot(components, cumulative_var, marker='o', color='orange', label='Cumulative Explained Variance', linewidth=2)

for i, val in enumerate(cumulative_var):
    plt.text(components[i], val + 0.02, f"{val:.2f}", ha='center', fontsize=8)

plt.xticks(components)
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.title('Explained Variance by Principal Components: Global Mental Health Dataset')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

Interpretation:

```python
# Elbow Method (for K-Means tuning)

# Purpose: Determine optimal number of clusters for K-Means.

k_range = range(1, 11)
inertia = []

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o')
plt.xticks(k_range)
plt.title('Elbow Plot: \nOptimal Clusters for K-Means')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Interpretation:
- There’s a noticeable “elbow” at k = 4, where the inertia reduction starts leveling off, suggesting that 4 clusters is a good balance between performance and overfitting.

- Going beyond k = 4 results in diminishing returns in clustering precision.

```python
# Silhouette Score (for cluster validation)

# Purpose: Evaluation of how well-separated each k cluster set is.

silhouette_scores = []

for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, labels))

# Plot
plt.figure(figsize=(8, 6))
plt.plot(range(2, 10), silhouette_scores, marker='o', color='deeppink')
plt.title('Silhouette Score for K-Means Clustering')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Interpretation:

- The highest silhouette score occurs at k = 4, confirming the elbow plot result.

- A silhouette score above 0.25 in unsupervised social science data is acceptable, and the best score of this plot being above 0.25, indicates that it is solid for real-world, noisy data.
